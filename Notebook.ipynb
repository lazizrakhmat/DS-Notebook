pandas kutubxonasi bilan ishlaydigan, fayllarni o'qish uchun funksiyalar:


* `read_csv` - vergul bilan ajratilgan files uchun
* `read_table` - probel (bo'sh joy) files uchun

* `read_clipboard` - ko'chirilgan (xotiradagi) jadvalni o'qish uchun
* `read_hdf` - HDF5 formatdagi fayllar uchun
* `read_html` HTML (web) formatdagi fayllar uchun
* `read_json` JSON ma'lumotlar o'qish
* `read_pickle` pickle formatidagi fayllar uchun
* `read_sql` SQL query dan qaytgan ma'lumotlarni o'qish (SQLAIchemy bilan ishlaydi)
